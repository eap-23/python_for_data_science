# Matplotlib - Part 3

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00', linewidth=3, alpha=0.2) #RGB Hex Code, alpha refers to transparency
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', linewidth=3, linestyle='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green', linewidth=1, linestyle='-', marker='o', markersize=10,
       markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='purple')
```

### Control Over Axis Appearance and Plot Range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='red', linewidth=2, linestyle='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

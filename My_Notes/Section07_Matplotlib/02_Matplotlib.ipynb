# Matplotlib - Part 2

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
#plt.subplots same as plt.figure except
#plt.subplots does the add_axes step when you designate rows and columns

fig, axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
#This is an array of axes based on tuple assignment above
axes
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[1].set_title('Second Plot')

axes[0].plot(x,y)
axes[0].set_title('First Plot')
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)

#plt.tight_layout will take care of any overlapping of axes
plt.tight_layout()
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_piture.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

#Can you use loc parameter to designate where to put legend in plot
#Can pass loc set code. Example 0 = best, 10 = center
#OR
#Can pass a tuple to loc to put legend at specific coordinates on figure
ax.legend()

#ax.legend(loc=10)
#ax.legend(loc=(0.1,0.1))
```

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#By default the estimator object for barplot is mean
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
#Estimator is type of aggregation
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
#Counts number of occurences in dataframe
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day', y='total_bill',data=tips)
```

```python
sns.violinplot(x='day', y='total_bill',data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill',data=tips, hue='sex', split=True)
```

```python
#Stripplot is scatter plot based off of the category
#jitter set to True by default
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
#swarmplot is like stripplot but points are adjusted so they don't overlap
#swarmplots not recommended for very large datasets
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
#factorplot is general form of all the above plots
#call what plot you want in the kind argument
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

## Final Thought
+ Know your audience
    + Boxplots and barplots great for presenting data to management
    + Violin plots or strip plots more advanced, and might be better for presenting to other data science specialists or for just exploring your data
